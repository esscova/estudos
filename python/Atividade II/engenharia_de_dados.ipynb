**Exercício 4.1:**
Carregue o dataset `titanic` e realize as seguintes transformações:

- Crie uma nova coluna `family_size` que seja a soma de `sibsp` (irmãos/cônjuges) e `parch` (pais/filhos).
- Converta a coluna `sex` em valores numéricos (0 para masculino e 1 para feminino).
- Preencha os valores ausentes na coluna `age` com a mediana da idade.

**Exercício 4.2:**
Crie uma função que receba um DataFrame e:

- Padronize as colunas numéricas (subtraia a média e divida pelo desvio padrão).
- Aplique codificação one-hot em colunas categóricas.

```python
# bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# dados
titanic = sns.load_dataset('titanic')
```

## TITANIC

```python
titanic.head()
```

```python
# Crie uma nova coluna family_size que seja a soma de sibsp (irmãos/cônjuges) e parch (pais/filhos).
titanic['family_size'] = titanic['sibsp'] + titanic['parch']
titanic.head()
```

```python
# Converta a coluna sex em valores numéricos (0 para masculino e 1 para feminino).
titanic['sex'] = titanic['sex'].map({'male': 0, 'female': 1})
titanic.head()
```

```python
titanic.sex.value_counts()
```

```python
# Preencha os valores ausentes na coluna age com a mediana da idade.
titanic['age'].fillna(titanic['age'].median(), inplace=True)
titanic.head()
```

```python
titanic.age.isna().any()
```

```python
# função que padronize as colunas numéricas (subtraia a média e divida pelo desvio padrão).
def padronizar_colunas_numericas(df):
    colunas_numericas = df.select_dtypes(include=['float64', 'int64']).columns
    df[colunas_numericas] = (df[colunas_numericas] - df[colunas_numericas].mean()) / df[colunas_numericas].std()
    return df

# função que aplica codificação one-hot em colunas categóricas.
def codificar_one_hot(df):
    colunas_categoricas = df.select_dtypes(include=['object']).columns
    df_codificado = pd.get_dummies(df, columns=colunas_categoricas, drop_first=True)
    return df_codificado
```

```python
df_padronizado = padronizar_colunas_numericas(titanic)
df_padronizado.head()
```

```python
df_padronizado_e_codificado = codificar_one_hot(df_padronizado)
df_padronizado_e_codificado.head()
```
