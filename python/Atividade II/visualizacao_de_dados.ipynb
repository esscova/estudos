- Um heatmap da matriz de correlação entre as variáveis numéricas.
- Um gráfico de violino para comparar a distribuição de idades por classe de cabine.

```python
# bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# dados
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
```

## IRIS

**Exploração**

```python
iris.head()
```

```python
iris.info()
```

```python
iris.describe()
```

### Um histograma para a distribuição de comprimento das pétalas (petal_length).

```python
sns.histplot(iris['petal_length'], kde=True);
```

### Um gráfico de dispersão (`scatter plot`) entre comprimento e largura das pétalas.

```python
sns.scatterplot(x='petal_length', y='petal_width', data=iris);
```

### Um boxplot para comparar a distribuição de comprimento das sépalas (`sepal_length`) por espécie.

```python
sns.boxplot(x='species', y='sepal_length', data=iris);
```

### Um gráfico de barras mostrando a contagem de sobreviventes vs. não sobreviventes.

```python
sns.countplot(x='survived', data=titanic)
plt.xlabel('Sobrevivente (0 = Não, 1 = Sim)')
plt.show()
```

### Um heatmap da matriz de correlação entre as variáveis numéricas.

```python
numericas = titanic.select_dtypes(include=np.number)
numericas.head()
```

```python
cor = numericas.corr()
cor
```

```python
mask = np.triu(np.ones_like(cor, dtype=bool))
```

```python
sns.heatmap(cor, mask=mask, annot=True, cmap='coolwarm')
plt.show()
```

### Um gráfico de violino para comparar a distribuição de idades por classe de cabine.

```python
sns.violinplot(x='class', y='age', data=titanic);
```

```python
sns.violinplot(x=titanic['class'], y=titanic['age'].fillna(titanic.age.mean()));
```

```python
sns.violinplot(x='class', y='age', hue='sex', data=titanic);
```

```python
sns.violinplot(x=titanic['class'], y=titanic['age'].fillna(titanic.age.mean()), hue=titanic['sex']);
```

```python
sns.violinplot(x='class', y='age', hue='survived', data=titanic);
```

```python
sns.violinplot(x=titanic['class'], y=titanic['age'].fillna(titanic.age.mean()), hue=titanic.survived);
```
