## Detector de faces com openCV

```python
import cv2
from google.colab.patches import cv2_imshow

def detect_faces(imagem:str, scale:float, minNeighbors:int):
    # classificador
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

    # classificador?
    if face_cascade.empty():
        print("Erro: Não foi possível carregar o Haar Cascade")
        exit()

    # carregar a imagem
    image = cv2.imread(imagem)
    if image is None: # imagem?
        print("Erro: Não foi possível carregar a imagem")
        exit()

    # grayscale e equalizacao do hist
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # detectando faces
    faces = face_cascade.detectMultiScale(gray, scale, minNeighbors, minSize=(30, 30))
    print(f"Número de rostos detectados: {len(faces)}")

    # marcando os rostos detectados
    for (x, y, w, h) in faces:
        cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)

    # resultado
    cv2_imshow(image)  # cv2.imshow() fora colab
    cv2.waitKey(0)
    cv2.destroyAllWindows()

try:
    detect_faces('/content/imagem.png', 1.3, 5)
except Exception as e:
    print(f'Erro: {e}')
```
