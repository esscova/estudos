```r
# carregando dados
data(iris)
```

```r
# primeiras linhas
head(iris)
```

```r
# matriz de distancia
dist_iris <- dist(iris[, -5])
```

```r
# agrupamento aglomerativo
hc <- hclust(dist_iris, method = "average")
```

```r
# plot dendograma
plot(hc)
```

```r
# obter 3 clusters
clusters <- cutree(hc, k=3)
```

```r
# adicionando clusters
plot(hc)
rect.hclust(hc, k=3, border=2:4)
```
