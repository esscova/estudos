# Agrupamento com K-Means em R

O aprendizado não supervisionado por agrupamento tem como objetivo **identificar grupos (clusters) de dados similares entre si, sem a necessidade de informações prévias sobre as classes**. O algoritmo **K-means** é um exemplo de método de agrupamento que se baseia em distâncias para encontrar os grupos.

```r
# carregando dados
data(iris)
```

```r
# kmeans
k <- 3
agrupamento <- kmeans(iris[, 1:4], centers = k)
```

```r
# clusters
clusters <- agrupamento$cluster
table(iris$Species, clusters)
```

```r
# adicionando clusters ao dataset
iris$cluster <- clusters
head(iris)
```

```r
# visualizando agrupamento
plot(iris[, 1:4], col=clusters)
```

```r
plot(iris$Sepal.Length, iris$Sepal.Width, col=clusters, pch = 20, main="iris")
```
