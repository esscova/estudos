# Clustering com Fuzzy C-Means com Python no dataset IRIS

O Fuzzy C-Means (FCM) é um algoritmo de agrupamento (clustering) que permite que um ponto de dados pertença a mais de um cluster simultaneamente, com diferentes graus de pertinência.

```python
# bibliotecas
import numpy as np
import matplotlib.pyplot as plt

from sklearn import datasets
!pip install scikit-fuzzy -q
from skfuzzy.cluster import cmeans
```

```python
# carregando dados
iris = datasets.load_iris()
data = iris.data.T # transpor linhas x colunas
```

```python
# modelo
cntr, u, u0, d, jm, p, fpc = cmeans(data, c=3, m=2, error=1e-5, maxiter=1000, seed=42)
```

```python
# centroides
cntr
```

```python
# medida de qualidade dos clusters
fpc
```

```python
# matriz de pertinencia
u
```

```python
# maior pertinencia ao cluster
cluster_membros = np.argmax(u, axis=0)
cluster_membros
```

```python
# plot
plt.scatter(data[0,:], data[1,:], c=cluster_membros, cmap='viridis', s=50, alpha=0.6)
plt.scatter(cntr[:,0], cntr[:,1], marker='o', s=200, c='red', label='Centroides')
plt.title('Fuzzy C-Means clustering')
plt.xlabel('Sepal length (cm)')
plt.ylabel('Setal width (cm)')
plt.legend()
plt.show()
```
