```r
# bibliotecas
library(cluster)
```

```r
# carregando dados
data(iris)
```

```r
# primeiras linhas
head(iris)
```

```r
# algoritmo k-medoids com k=3 excluindo atributo species
kmedoids <- pam(iris[, -5], k=3)
```

```r
# resultados
print(kmedoids)
```

```r
# plot
plot(iris[,-5], col=kmedoids$clustering)
points(kmedoids$medoids, col=1:3, pch=8, cex=2)
```
